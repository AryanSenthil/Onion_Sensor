## <span style="color: rgb(2, 14, 26); border-bottom: 2px solid #3498db;">1. Piezoelectricity and Piezoelectric Effect</span>

<table style="width: 100%; border: none;">
<tr>
<td style="width: 60%; vertical-align: top; padding-right: 20px; border: none;">

<div style="background-color: #f8f9fa; padding: 12px 15px; border-left: 3px solid #3498db; margin: 10px 0;">
<strong>Piezoelectricity</strong> is the linear electromechanical coupling between mechanical and electrical domains.<br><br>
<strong>Piezoelectric effect</strong> refers to an electric polarization induced by an applied mechanical stress (σ).
</div>

### <span style="color: #3498db;">1.1 Mathematical Expression</span>
<div style="background-color: #ecf0f1; padding: 15px; border-radius: 5px; text-align: center; margin: 8px 0;">

$$ P_i = d_{ijk} \, \sigma_{jk} $$
</div>

<div style="background-color: #fff; padding: 10px 15px; border: 1px solid #ddd; border-radius: 5px; margin: 8px 0;">

<strong>where:</strong> $P_i$ = Electric polarization, $d_{ijk}$ = Piezoelectric coefficient tensor, $\sigma_{jk}$ = Applied mechanical stress
</div>

### <span style="color: #3498db;">1.2 Why Piezoelectric Materials?</span>

<div style="background-color: #e8f4f8; padding: 12px 15px; border-left: 3px solid #2980b9; margin: 10px 0;">
For <strong>piezoelectric materials</strong>, the crystal's structure is <strong>asymmetrical</strong>. When no external force is applied, the crystal is in a stable state and doesn't display electrical polarity to the outside. However, when an external force is applied, the crystal's atoms disperse and the internal charge distribution shifts, creating <strong>electrical polarity</strong> as the balance is disrupted.
</div>

### <span style="color: #d32f2f;">2.1 Material Selection and Rationale</span>

<div style="background-color: #fdf5f5; padding: 12px 15px; border-left: 3px solid #a94442; margin: 10px 0;">

In this research, we explored **onion peel** as a sustainable piezoelectric material due to its self-aligned cellulose fibers that exhibit polarization under applied mechanical stress. This approach transforms agricultural waste into a functional sensing material.

</div>

### <span style="color: #d32f2f;">2.2 Material Preparation Protocol</span>

<div style="background-color: #fff; padding: 12px 15px; border: 1px solid #d9a5a5; border-radius: 5px; margin: 10px 0;">

1. **Collection and Dehydration**: Onion peel waste material was collected and thoroughly dehydrated
2. **Grinding**: Dehydrated peels were ground using a high-speed grinder (3600 rpm)
3. **Sieving**: Ground material was sieved using a vibrating sieve shaker to obtain uniform particle size distribution (roughly 25 $\,\mu\text{m}$)

</div>

### <span style="color: #d32f2f;">2.3 Sensor Fabrication Process</span>

<div style="background-color: #f9eded; padding: 12px 15px; border-left: 3px solid #c9302c; margin: 10px 0;">

**Step 1: Composite Ink Preparation**
- Onion peel powder was dispersed in a silicone matrix (Ecoflex 00-30, 1:1 base:curing agent ratio)
- The mixture was homogenized at 1500 rpm for 5 minutes to form an active composite ink

**Step 2: 3D Printing via Direct Ink Writing (DIW)**
- The composite ink was printed layer-by-layer using a pneumatic DIW printer
- Ink rheology was carefully tuned to ensure smooth extrusion and dimensional stability during printing

**Step 3: Electrode Application**
- Conductive electrodes were applied to both surfaces of the printed sensor using silver paste to establish electrical contacts

</div>

### <span style="color: #d32f2f;">2.4 Characterization and Testing</span>

<div style="background-color: #fff9f9; padding: 12px 15px; border-left: 3px solid #d9534f; margin: 10px 0;">

The fabricated sensor was connected to an **electrometer** to measure voltage generation under mechanical loading. Two loading conditions were evaluated:

- **Point Load**: Localized force applied to a specific region of the sensor
- **Uniform Load**: Distributed force applied across the entire sensor surface

As expected, voltage output was successfully measured under both loading conditions, confirming the piezoelectric response of the onion peel-based sensor.

</div>

Piezoelectricity is a function of stress. The mechanical stress (σ) applied to the sensor is related to the deformation through Hooke's law:

<div style="background-color: #fafafa; padding: 15px; border-radius: 5px; text-align: center; margin: 10px 0;">

</div>

# <span style="color:rgb(13, 1, 21); border-bottom: 3px solid #6A1B9A;">4. Machine Learning Model Architecture for Load Characterization and Deformation Prediction</span>

<div style="background-color: #F3E5F5; padding: 15px 20px; border-left: 4px solid #6A1B9A; margin: 15px 0; border-radius: 5px;">
The piezoelectric response of our onion peel-based sensor exhibits distinct voltage-time signatures that correlate strongly with both the type of applied load (point vs. uniform) and the magnitude of deformation. We developed a novel <strong>two-stage machine learning architecture</strong> that leverages these correlations to simultaneously classify load type and predict deformation level from the sensor's electrical output.
</div>

## <span style="color: #6A1B9A;">4.1 Two-Stage Model Architecture</span>

## <span style="color: #6A1B9A;">4.2 Load Classification Module</span>

<div style="background-color: #F3E5F5; padding: 15px 20px; border-left: 4px solid #6A1B9A; margin: 15px 0; border-radius: 5px;">
The cyclic mechanical loading produces characteristic voltage waveforms that vary distinctly between point and uniform load configurations. To classify these patterns, we treat the voltage-time data as audio signals and employ <strong>convolutional neural networks (CNNs)</strong> on their time-frequency representations.
</div>

### <span style="color: #6A1B9A;">4.2.1 Signal Processing Pipeline</span>

<div style="background-color: #fff; padding: 12px 18px; border: 2px solid #6A1B9A; border-radius: 5px; margin: 10px 0;">
<strong style="color: #6A1B9A;">Data Segmentation:</strong> Continuous voltage-time recordings are segmented into 10-second intervals, each saved as a WAV file<br><br>
<strong style="color: #6A1B9A;">Time-Frequency Analysis:</strong> We apply the Short-Time Fourier Transform (STFT) to generate spectrograms
</div>

### <span style="color: #6A1B9A;">4.2.2 Mathematical Expression</span>

<table style="width: 100%; border: none;">
<tr>
<td style="width: 60%; vertical-align: top; padding-right: 20px; border: none;">

<div style="background-color: #E1BEE7; padding: 20px; border-radius: 8px; text-align: center; margin: 15px 0;">

$$ X(\tau, \omega) = \int_{-\infty}^{\infty} x(t) \cdot w(t - \tau) \cdot e^{-j\omega t} \, dt $$

</div>

<div style="background-color: #fff; padding: 12px 18px; border: 2px solid #6A1B9A; border-radius: 5px; margin: 10px 0;">

<strong style="color: #6A1B9A;">where:</strong> $x(t)$ = voltage signal, $w(t-\tau)$ = window function (Hann window), $\tau$ = time frame, $\omega$ = angular frequency
</div>

</td>

### <span style="color: #6A1B9A;">4.2.3 Why STFT for Cyclic Loading Analysis?</span>

<div style="background-color: #F3E5F5; padding: 18px 22px; border-left: 5px solid #6A1B9A; border-radius: 5px; margin: 15px 0; box-shadow: 0 3px 10px rgba(106,27,154,0.25);">
<strong style="color: #6A1B9A;">Non-stationary Signal Characteristics:</strong> The voltage response during cyclic compression-decompression is inherently non-stationary, the frequency content changes as the sensor transitions between loading and unloading phases. STFT preserves temporal evolution, capturing how spectral components emerge and decay during each loading cycle.
</div>

<div style="background-color: #F3E5F5; padding: 18px 22px; border-left: 5px solid #6A1B9A; border-radius: 5px; margin: 15px 0; box-shadow: 0 3px 10px rgba(106,27,154,0.25);">
<strong style="color: #6A1B9A;">Cycle-Specific Features:</strong> Point and uniform loads produce different stress distributions across the sensor, resulting in distinct harmonic patterns during compression. STFT reveals these time-localized frequency signatures. Point loads generate sharper transients with higher frequency components at the moment of contact, while uniform loads show more gradual spectral evolution.
</div>

<div style="background-color: #F3E5F5; padding: 18px 22px; border-left: 5px solid #6A1B9A; border-radius: 5px; margin: 15px 0; box-shadow: 0 3px 10px rgba(106,27,154,0.25);">
<strong style="color: #6A1B9A;">Temporal Resolution:</strong> Our 10-second recording windows contain multiple compression cycles. STFT allows us to resolve individual cycles within this window, providing a detailed time-frequency map that the CNN can learn to associate with specific load types.
</div>

<div style="background-color: #F3E5F5; padding: 18px 22px; border-left: 5px solid #6A1B9A; border-radius: 5px; margin: 15px 0; box-shadow: 0 3px 10px rgba(106,27,154,0.25);">
<strong style="color: #6A1B9A;">Robustness to Phase Variations:</strong> Since mechanical loading may not be perfectly synchronized across samples, STFT's sliding window approach provides translation-invariant features that are robust to slight timing variations in the compression cycles.
</div>

## <span style="color: #6A1B9A;">4.3 Deformation Prediction Module: Wide & Deep Regression</span>

<div style="background-color: #F3E5F5; padding: 15px 20px; border-left: 4px solid #6A1B9A; margin: 15px 0; border-radius: 5px;">
Initial experiments using only peak-to-peak voltage as a predictor for deformation showed promising results. However, incorporating the load type as an additional feature in traditional multi-layer perceptron architectures yielded minimal improvement, the categorical nature of the load type was being diluted during training.
</div>

<div style="background-color: #F3E5F5; padding: 18px 22px; border-left: 5px solid #6A1B9A; border-radius: 5px; margin: 15px 0; box-shadow: 0 3px 10px rgba(106,27,154,0.25);">
We adapted the <strong><a href="https://arxiv.org/pdf/1606.07792" target="_blank">Wide & Deep architecture</a></strong> (Cheng et al., 2016), originally developed for recommender systems. This architecture elegantly combines memorization and generalization.
</div>

### <span style="color: #6A1B9A;">4.3.1 Architecture Components</span>

<div style="background-color: #fff; padding: 12px 18px; border: 2px solid #6A1B9A; border-radius: 5px; margin: 10px 0;">
<strong style="color: #6A1B9A;">Wide Component:</strong> A linear model that directly connects the load type (categorical feature) to the output, preserving its discriminative power through memorization<br><br>
<strong style="color: #6A1B9A;">Deep Component:</strong> A deep neural network that processes the peak-to-peak voltage through multiple non-linear transformations, enabling generalization across the continuous deformation space
</div>

## <span style="color: #6A1B9A;">4.4 Integration and Inference Pipeline</span>

<div style="background-color: #F3E5F5; padding: 15px 20px; border-left: 4px solid #6A1B9A; margin: 15px 0; border-radius: 5px;">
During inference, the system operates as follows:<br><br>
<strong style="color: #6A1B9A;">1.</strong> Input voltage-time data is processed through the CNN classifier, outputting load type probabilities<br>
<strong style="color: #6A1B9A;">2.</strong> The predicted load type (argmax of probabilities) is converted to binary representation<br>
<strong style="color: #6A1B9A;">3.</strong> Peak-to-peak voltage is extracted from the raw signal<br>
<strong style="color: #6A1B9A;">4.</strong> Both features feed into the Wide & Deep regression model for deformation prediction
</div>

<div style="background-color: #FFF3E0; padding: 15px 20px; border-left: 4px solid #FF9800; margin: 15px 0; border-radius: 5px;">
<strong style="color: #FF9800;">Note on Feature Representation:</strong> While the CNN outputs a probability distribution over load types during inference, the regression model uses a discrete binary input (0 or 1) to maintain consistency with the training paradigm where ground truth labels were used.
</div>

# Implementation

The following cells contain the executable code that implements the methodology described above.

## Setup and Configuration

```python
import pathlib
from pathlib import Path
import json
import numpy as np
import tensorflow as tf

import classification_module 
import regression_module
from tensorflow import keras

%load_ext autoreload
%autoreload 2
```

```python
# Configuration
classification_model_name = "classification_model"
regression_model_name = "regression_model"

CSV_DIR = r"/home/ari/Documents/Onion_Sensor/data"
SAMPLING_RATE = 1600
VALIDATION_SPLIT = 0.2
LEARNING_RATE = 0.001
Classification_EPOCHS = 10  
REGRESSION_EPOCHS = 3000
PATIENCE = 2
SEED = 42
TEST_SPLIT = 0.1
```

## Classification Module Training

```python
# Step 1: Preprocess CSV files to WAV format for STFT analysis
classification_module.preprocess_csv_to_wav(
    csv_dir=CSV_DIR,
    model_name=classification_model_name,
    sampling_rate=SAMPLING_RATE
)
```

```python
# Step 2: Load and preprocess spectrograms
classification_module.load_and_preprocess_data(
    model_name=classification_model_name,
    validation_split=VALIDATION_SPLIT,
    seed=SEED
)
```

```python
# Step 3: Build model
classification_module.build_and_compile_model(
    model_name=classification_model_name,
    learning_rate=LEARNING_RATE
)
```

```python
# Step 4: Train CNN classification model
classification_module.train_and_evaluate_model(
    model_name=classification_model_name,
    epochs=Classification_EPOCHS,
    patience=PATIENCE
)
```

```python
# Step 5: Export trained classification model
classification_module.create_export_model(model_name=classification_model_name)
```

## Regression Module Training

```python
# Step 1: Preprocess regression data (extract peak-to-peak voltage and load types)
regression_module.preprocess_csv_data(
    csv_dir=CSV_DIR,
    model_name=regression_model_name,
    test_split=TEST_SPLIT,
    seed=SEED
)
```

```python
# Step 2: Build model
regression_module.build_and_compile_model(
    model_name=regression_model_name,
    hidden_layers=[64, 64, 32, 16],
    learning_rate=LEARNING_RATE
)
```

```python
# Step 3: Train Wide & Deep regression model
regression_module.train_and_evaluate_model(
    model_name=regression_model_name,
    epochs=REGRESSION_EPOCHS,
    patience=5
)
```

```python
# Step 4: Export trained regression model
regression_module.create_export_model(model_name=regression_model_name)
```

## Inference Pipeline
