## <span style="color: rgb(2, 14, 26); border-bottom: 2px solid #3498db;">Piezoelectricity and Piezoelectric Effect</span>

<table style="width: 100%; border: none;">
<tr>
<td style="width: 60%; vertical-align: top; padding-right: 20px; border: none;">

<div style="background-color: #f8f9fa; padding: 12px 15px; border-left: 3px solid #3498db; margin: 10px 0;">
<strong>Piezoelectricity</strong> is the linear electromechanical coupling between mechanical and electrical domains.<br><br>
<strong>Piezoelectric effect</strong> refers to an electric polarization induced by an applied mechanical stress (σ).
</div>

### <span style="color: #3498db;">Mathematical Expression</span>
<div style="background-color: #ecf0f1; padding: 15px; border-radius: 5px; text-align: center; margin: 8px 0;">

$$ P_i = d_{ijk} \, \sigma_{jk} $$
</div>

<div style="background-color: #fff; padding: 10px 15px; border: 1px solid #ddd; border-radius: 5px; margin: 8px 0;">

<strong>where:</strong> $P_i$ = Electric polarization, $d_{ijk}$ = Piezoelectric coefficient tensor, $\sigma_{jk}$ = Applied mechanical stress
</div>

### <span style="color: #3498db;"> Why Piezoelectric Materials?</span>

<div style="background-color: #e8f4f8; padding: 12px 15px; border-left: 3px solid #2980b9; margin: 10px 0;">
For <strong>piezoelectric materials</strong>, the crystal's structure is <strong>asymmetrical</strong>. When no external force is applied, the crystal is in a stable state and doesn't display electrical polarity to the outside. However, when an external force is applied, the crystal's atoms disperse and the internal charge distribution shifts, creating <strong>electrical polarity</strong> as the balance is disrupted.
</div>

## <span style="color:rgb(2, 14, 26); border-bottom: 2px solid #d32f2f;">Fabrication of Onion Peel-Based Piezoelectric Sensor</span>

### <span style="color: #d32f2f;">Material Selection and Rationale</span>

<div style="background-color: #fdf5f5; padding: 12px 15px; border-left: 3px solid #a94442; margin: 10px 0;">

In this research, we explored **onion peel** as a sustainable piezoelectric material due to its self-aligned cellulose fibers that exhibit polarization under applied mechanical stress. This approach transforms agricultural waste into a functional sensing material.

</div>

### <span style="color: #d32f2f;">Material Preparation Protocol</span>

<div style="background-color: #fff; padding: 12px 15px; border: 1px solid #d9a5a5; border-radius: 5px; margin: 10px 0;">

1. **Collection and Dehydration**: Onion peel waste material was collected and thoroughly dehydrated
2. **Grinding**: Dehydrated peels were ground using a high-speed grinder (3600 rpm)
3. **Sieving**: Ground material was sieved using a vibrating sieve shaker to obtain uniform particle size distribution (roughly 25 $\,\mu\text{m}$)

</div>

### <span style="color: #d32f2f;">Sensor Fabrication Process</span>

<div style="background-color: #f9eded; padding: 12px 15px; border-left: 3px solid #c9302c; margin: 10px 0;">

**Step 1: Composite Ink Preparation**
- Onion peel powder was dispersed in a silicone matrix (Ecoflex 00-30, 1:1 base:curing agent ratio)
- The mixture was homogenized at 1500 rpm for 5 minutes to form an active composite ink

**Step 2: 3D Printing via Direct Ink Writing (DIW)**
- The composite ink was printed layer-by-layer using a pneumatic DIW printer
- Ink rheology was carefully tuned to ensure smooth extrusion and dimensional stability during printing

**Step 3: Electrode Application**
- Conductive electrodes were applied to both surfaces of the printed sensor using silver paste to establish electrical contacts

</div>

### <span style="color: #d32f2f;">Characterization and Testing</span>

<div style="background-color: #fff9f9; padding: 12px 15px; border-left: 3px solid #d9534f; margin: 10px 0;">

The fabricated sensor was connected to an **electrometer** to measure voltage generation under mechanical loading. Two loading conditions were evaluated:

- **Point Load**: Localized force applied to a specific region of the sensor
- **Uniform Load**: Distributed force applied across the entire sensor surface

As expected, voltage output was successfully measured under both loading conditions, confirming the piezoelectric response of the onion peel-based sensor.

## <span style="color:rgb(2, 14, 26); border-bottom: 2px solid #ff6f00;">Experimental Data Collection</span>

<table style="width: 100%; border: none;">
<tr>
<td style="width: 100%; vertical-align: top; border: none;">

<div style="background-color: #fff; padding: 12px 15px; border: 1px solid #bdbdbd; border-radius: 5px; margin: 10px 0; color: #000;">

Piezoelectricity is a function of stress. The mechanical stress (σ) applied to the sensor is related to the deformation, i.e., change in sensor height, through Hooke's law:

<div style="background-color: #fafafa; padding: 15px; border-radius: 5px; text-align: center; margin: 10px 0;">

$$\mathbf{\sigma} = \mathbf{E} \cdot \frac{\mathbf{\Delta h}}{\mathbf{h_0}}$$

</div>

<strong style="color: #000;">where:</strong> $\mathbf{E}$ = Young's modulus of the composite material, $\mathbf{\Delta h}$ = deformation, i.e., change in sensor height, $\mathbf{h_0}$ = original sensor height (2.64 mm)

<div style="background-color: #fff3e0; padding: 15px 20px; border-left: 5px solid #ff6f00; border-radius: 5px; margin: 15px 0; box-shadow: 0 2px 8px rgba(255,111,0,0.2); color: #000;">
<strong style="color: #ff6f00; font-size: 1.05em;">Experimental Approach:</strong><br>
The piezoelectric response of the onion peel-based sensor is characterized through <strong>cyclic compression testing at three deformation levels</strong> (Δh = 0.5, 0.7, and 1.0 mm) under <strong>two distinct loading configurations</strong> (point and uniform load), with continuous electrical measurements of <strong> Voltage (V) Vs. Time (s) </strong> captured throughout each loading cycle.
</div>

</div>

<span style="color: #ff6f00; font-weight: bold;">Sensor Geometry:</span> Original height ($h_0$) = 2.64 mm
